```python
df = "C:/Users/adars/Downloads/Resources/avg_rain_state.csv"
pd.read_csv(df)

df = pd.read_csv(df)

df.head(5)
```

```python
tick_locations = [value for value in df["State"]]
plt.figure(figsize=(20,3))
plt.bar(df["State"],df["Inches"])
plt.xticks(tick_locations, df["State"], rotation= 'horizontal')
plt.title("Rain in Inches by State")
plt.xlabel("State")
plt.ylabel("Rain (inches)")
plt.show()
```

```python
#For the following data. 

test_grades = {
    'Class': ['Oct', 'Oct', 'Jan', 'Jan', 'Oct', 'Jan'], 
    'Name': ["Cyndy", "Logan", "Laci", "Elmer", "Crystle", "Emmie"], 
    'Test Score': [90, 59, 72, 88, 98, 60]}

#Create a DataFrame from the “test_grades” dictionary. 
#Create bins, 0, 59, 69, 79, 89, 100, to hold the data.
#Assign a letter grade, as “letter_grades”, for the four bins as the labels.
#Complete the code that will create a new column “Letter Grade” based on the four bins and labels
df = pd.DataFrame(test_grades)
bins = [0, 59, 69, 79, 89, 100]
letter_grades = ["F (<60)", "D (60 - 69)", "C (70 - 79)", "B (80-89)", "A (90 - 100)"]

df['letter_grades'] = pd.cut(df['Test Score'], bins, labels = letter_grades)
```
